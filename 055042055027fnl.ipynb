## Description of the Data
This dataset provides detailed information on international trade transactions, capturing both import and export activities. It includes comprehensive data on various aspects of trade, making it a valuable resource for business analysis, economic research, and financial modeling.

Features:

Transaction_ID: Unique identifier for each trade transaction.
Country: Country of origin or destination for the trade.
Product: Product being traded.
Import_Export: Indicates whether the transaction is an import or export.
Quantity: Amount of the product traded.
Value: Monetary value of the product in USD.
Date: Date of the transaction.
Category: Category of the product (e.g., Electronics, Clothing, Machinery).
Port: Port of entry or departure.
Customs_Code: Customs or HS code for product classification.
Weight: Weight of the product in kilograms.
Shipping_Method: Method used for shipping (e.g., Air, Sea, Land).
Supplier: Name of the supplier or manufacturer.
Customer: Name of the customer or recipient.
Invoice_Number: Unique invoice number for the transaction.
Payment_Terms: Terms of payment (e.g., Net 30, Net 60, Cash on Delivery).
Usage: This dataset can be used for various analyses, including:

Trade Analysis: Understanding trade patterns and trends.
Economic Research: Studying the impact of trade on different economies.
Financial Modeling: Evaluating financial performance and risks associated with trade.
Supply Chain Management: Analyzing logistics and shipping methods.
Data Format: The dataset is provided in CSV format, making it easy to load into data analysis tools and frameworks.

Additional Notes:
The dataset consists of 15,000 rows, ensuring a substantial sample size for meaningful analysis.
Data is generated with realistic variability, simulating actual trade transactions.

### Test of Normality (Shapiro-Wilk Test): Managerial Insights

**Inventory Management**: If the weight of products is not normally distributed, it suggests a wide variation in the weight of items being imported or exported. This information is crucial for inventory management, impacting warehouse space allocation, storage requirements, and handling procedures. Managers need to consider this variability when planning storage capacity and optimizing warehouse layouts.

**Shipping Costs**: Non-normality in weight distribution can significantly influence shipping costs. Heavier items generally cost more to transport, and a skewed distribution might indicate a higher proportion of heavier products. Managers should analyze the weight distribution to accurately estimate shipping expenses and explore strategies to minimize costs, such as negotiating bulk rates for heavier shipments or optimizing packaging to reduce weight.

**Logistics and Transportation**: The weight distribution affects various aspects of logistics and transportation, including vehicle selection, route planning, and load balancing. A non-normal distribution might require specialized transportation solutions for heavier or oversized items. Managers should consider these factors when selecting transportation partners and planning delivery routes.

**Risk Assessment**: Non-normality can have implications for risk assessment, particularly concerning damage or loss during shipping. If a significant portion of products falls within a specific weight range, there might be a higher risk of damage for those items. Managers should identify potential risks associated with the weight distribution and implement appropriate mitigation strategies.

**Demand Forecasting**: Understanding the distribution of product weights can be valuable for demand forecasting. If certain weight categories are more prevalent, it suggests higher demand for those types of products. Managers can use this information to adjust inventory levels and anticipate future demand patterns more accurately.

### Recommendations

#### Test of Proportion (Chi-square Test of Independence)

```python
#Chi-square Test for Proportions of Import vs Export in Shipping Methods
import pandas as pd  # Import pandas
contingency_table = pd.crosstab(my_sample['Import_Export'], my_sample['Shipping_Method']) 

# Perform Chi-square test
from scipy.stats import chi2_contingency  # Import chi-square test function
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}, P-value: {p}")
```

- Since the p-value is greater than 0.05, we fail to reject the null hypothesis.
- The lack of a statistically significant association suggests that the choice of shipping method (Air, Land, Sea) is not influenced by 
whether the transaction is an import or export.  This could indicate that the same shipping methods are viable for both importers and 
exporters,leading to operational flexibility.
- Suggestion: Foster closer relationships with suppliers and carriers, focusing on those who provide the most frequently used shipping methods. Explore opportunities for negotiated rates or value-added services.

#### Test of Normality (Shapiro-Wilk Test)

```python
# Shapiro-Wilk Test for Normality of Weight
from scipy.stats import shapiro
stat, p_value = shapiro(my_sample['Weight'])
print(f"Shapiro-Wilk Test Statistic: {stat}, P-value: {p_value}")
```

- Since the p-value is significantly less than 0.05, we reject the null hypothesis.
- This means that there is strong evidence to suggest that the data is not normally distributed.

#### Test of Variance (Levene’s / Bartlett’s Test)

```python
# Levene’s Test for equal variances between Exports and Imports
from scipy.stats import levene
exports = my_sample[my_sample['Import_Export'] == 'Export']['Value']
imports = my_sample[my_sample['Import_Export'] == 'Import']['Value']
stat, p = levene(exports, imports)
print(f"Levene's Test Statistic: {stat}, P-value: {p}")
```

- Since the p-value is greater than 0.05, we fail to reject the null hypothesis.
- This means that there is no significant difference in the variances between the groups being compared. In other words, the groups likely have equal variances.

#### Test of Correlation (Pearson Correlation)

```python
# Pearson Correlation between Value and Weight
correlation_value_weight = my_sample['Value'].corr(my_sample['Weight'])

print(f"Pearson Correlation between Value and Weight: {correlation_value_weight}")
```

- Since the correlation is very close to 0, there is almost no linear relationship between the transaction value and the weight of products in the dataset.
- In practical terms, changes in product weight have little to no impact on the transaction value.

#### Test of Mean (t-test / ANOVA)

```python
# t-test for Mean Difference in Value between Shipping Methods (e.g., Land vs Sea)
from scipy.stats import ttest_ind

land_value = my_sample[my_sample['Shipping_Method'] == 'Land']['Value']
sea_value = my_sample[my_sample['Shipping_Method'] == 'Sea']['Value']

t_stat, p_value = ttest_ind(land_value, sea_value)

print(f"T-statistic: {t_stat}, P-value: {p_value}")
```
